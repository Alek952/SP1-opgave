```python
import pandas as pd 

data = pd.read_csv("MoviesTop100.csv")
data.columns
display(data)
```

```python
#Spørgsmål 1
disney = data[data['Distributor'].str.contains('Disney') ==True]
```

```python
#Spørgsmål 1
disney['World Sales (in $)'].nlargest(n=10)
```

```python
#Spørgsmål 2

license = data['License'].unique()
distributers = data.iloc[:, 9]

PG = []
R = []
G = []

for x in distributers:
    if x == 'PG':
        PG.append(x)
    if x == 'R':
        R.append(x)
    if x == 'G':
        G.append(x)
    if x is None:
        pass
```

```python
#Spørgsmål 2

import matplotlib.pyplot as plt
import numpy as np

diffLicense = np.array([len(PG), len(R), len(G)])
myLabels = 'PG', 'R', 'G'

plt.pie(diffLicense, labels = myLabels)

plt.show()
```
